```python
from PIL import Image
from io import BytesIO

def convert_image_format(image_data, original_format, new_format):
    # Open the image from bytes
    image = Image.open(BytesIO(image_data))

    # Convert the image to the new format
    output_buffer = BytesIO()
    image.save(output_buffer, format=new_format)
    
    # Get the bytes data of the new image
    new_image_data = output_buffer.getvalue()

    return new_image_data
```

```python
with open('Deepika.jpg', 'rb') as file:
    jpg_data = file.read()
jpeg_data = convert_image_format(jpg_data, 'JPG','JPRG')
```
